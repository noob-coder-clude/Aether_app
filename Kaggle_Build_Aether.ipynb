# 🚀 Aether Android APK Automated Builder for Kaggle Notebooks

این نوت‌بوک برای ساخت فایل APK اپلیکیشن **Aether** در محیط کاگل (Kaggle Notebook) بهینه شده است.

```python
# Step 1: System setup for Kaggle Environment
!apt-get update -qq && apt-get install -y -qq openjdk-17-jdk git curl tar unzip

!if [ ! -d "/kaggle/working/flutter" ]; then \
  git clone https://github.com/flutter/flutter.git -b stable --depth 1 /kaggle/working/flutter; \
fi

import os
os.environ["PATH"] += ":/kaggle/working/flutter/bin"
os.environ["JAVA_HOME"] = "/usr/lib/jvm/java-17-openjdk-amd64"

!flutter --version
```

```python
# Step 2: Fetch Dependencies & Native Binaries
!chmod +x scripts/download_aether_binaries.sh
!./scripts/download_aether_binaries.sh
```

```python
# Step 3: Compile R8 Optimized Release APK
!flutter pub get
!flutter build apk --release

# Copy build artifact to output directory for easy download from Kaggle output tab
!cp build/app/outputs/flutter-apk/app-release.apk /kaggle/working/Aether-v1.2.0-release.apk
!ls -la /kaggle/working/Aether-v1.2.0-release.apk
```
