# 🚀 Aether Android APK Automated Builder for Google Colab

این نوت‌بوک جهت بیلد اتوماتیک و یک‌کلیکه برنامه **Aether Android Client** توسعه یافته توسط CluvexStudio طراحیشده است.

### ویژگی‌ها:
- نصب خودکار Java OpenJDK 17، Android SDK و Flutter SDK
- دانلود و بسته‌بندی هسته باینری نیتیو Aether برای تمامی معماری‌ها (`arm64`, `armv7`, `x86_64`)
- فعال‌سازی فشرده‌سازی **R8 & ProGuard** جهت ایجاد APK فوق سبک
- دانلود مستقیم فایل APK نهایی بعد از اتمام بیلد

```python
# Step 1: Install Dependencies (Java 17 & Flutter SDK)
!apt-get update -qq && apt-get install -y -qq openjdk-17-jdk curl git tar unzip wget
!export JAVA_HOME=/usr/lib/jvm/java-17-openjdk-amd64

# Download & Install Flutter SDK 3.22.x
!if [ ! -d "/opt/flutter" ]; then \
  git clone https://github.com/flutter/flutter.git -b stable --depth 1 /opt/flutter; \
fi

import os
os.environ["PATH"] += ":/opt/flutter/bin"
os.environ["JAVA_HOME"] = "/usr/lib/jvm/java-17-openjdk-amd64"

!flutter --version
```

```python
# Step 2: Download Aether Native Core Binaries & Prepare Workspace
!chmod +x scripts/download_aether_binaries.sh
!./scripts/download_aether_binaries.sh
```

```python
# Step 3: Build R8 Minified & Lightweight Release APK
!flutter pub get
!flutter build apk --release

# Show output APK files
!ls -lh build/app/outputs/flutter-apk/*.apk
```

```python
# Step 4: Download Release APK to Local Machine
from google.colab import files
import os

apk_path = 'build/app/outputs/flutter-apk/app-release.apk'
if os.path.exists(apk_path):
    print("Downloading Aether-release.apk...")
    files.download(apk_path)
else:
    print("APK build not found. Check compilation logs.")
```
